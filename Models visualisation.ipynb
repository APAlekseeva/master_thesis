### Concurrent tasks pattern

```python

from itertools import combinations

def find_concurrent_task_pairs(event_log):
    concurrent_task_pairs = []
    transitions = set(event_log['problem'])  # Get unique transition names
    
    # Check for concurrent task pairs
    for transition1, transition2 in combinations(transitions, 2):
        concurrent1 = event_log[event_log['problem'] == transition1]
        concurrent2 = event_log[event_log['problem'] == transition2]
        
        if len(concurrent1) > 0 and len(concurrent2) > 0:
            concurrent_task_pairs.append((transition1, transition2))
    
    return concurrent_task_pairs

# 'event_log' is a Pandas DataFrame
# 'problem' represents task names


# Find concurrent task pairs
concurrent_task_pairs = find_concurrent_task_pairs(event_log)
print("Concurrent Task Pairs:", concurrent_task_pairs)
```

### Dependent tasks pattern
#### Dependent tasks pairs:

```python
def find_dependent_task_pairs(event_log):
    dependent_task_pairs = set()  # Use set to store unique pairs
    
    transitions = set(event_log['problem'])  # Get unique transition names
    
    # Iterate over unique transition names
    for transition in transitions:
        task_indices = event_log.index[event_log['problem'] == transition].tolist()
        
        # Check if there are consecutive occurrences of the same task
        for i in range(len(task_indices) - 1):
            if task_indices[i + 1] - task_indices[i] == 1:
                dependent_task_pairs.add((transition, transition))  # Add unique pair to set
    
    return dependent_task_pairs

# 'event_log' is a Pandas DataFrame
# 'problem' represents task names


# Find unique pairs of dependent tasks
unique_dependent_task_pairs = find_dependent_task_pairs(event_log)
print("Unique Pairs of Dependent Tasks:", unique_dependent_task_pairs)
```

#### Dependent tasks sequences:

```python
def find_dependent_task_sequences(event_log):
    dependent_task_sequences = []  # Initialize list to store sequences
    
    transitions = event_log['problem'].tolist()  # Get list of task transitions
    
    # Iterate over the transitions
    i = 0
    while i < len(transitions) - 1:
        sequence = []  # Initialize sequence for each iteration
        
        # Check for consecutive dependent tasks
        if transitions[i] == transitions[i + 1]:
            task = transitions[i]  # Store the dependent task
            
            # Continue adding tasks to the sequence until a different task is encountered
            while i < len(transitions) and transitions[i] == task:
                sequence.append(task)
                i += 1
        else:
            i += 1  # Move to the next transition
        
        if sequence:
            dependent_task_sequences.append(tuple(sequence))  # Convert to tuple and append to the list
    
    return dependent_task_sequences

# 'event_log' is a Pandas DataFrame
# 'problem' represents task names

# Find unique sequences of dependent tasks
unique_dependent_task_sequences = find_dependent_task_sequences(event_log)
print("Unique Sequences of Dependent Tasks:", unique_dependent_task_sequences)
```
