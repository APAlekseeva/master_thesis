```python
import pandas as pd

def dataset_overview(dataframe):
    # Print the shape of the dataset
    print("Shape of the dataset:")
    print(dataframe.shape)
    print()

    # Display the first few rows of the dataset
    print("First few rows of the dataset:")
    print(dataframe.head())
    print()

    # Display basic statistics of numerical columns
    print("Basic statistics of numerical columns:")
    print(dataframe.describe())
    print()

    # Display information about the data types and non-null values
    print("Information about the dataset:")
    print(dataframe.info())
    print()

# Example usage:
# Assuming 'df' is your DataFrame
# Replace 'df' with the name of your DataFrame
dataset_overview(event_log)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the frequency of each event in the event log
event_counts = event_log['problem'].value_counts()
# Plotting a bar plot to visualize event frequencies
plt.figure(figsize=(10, 6))
sns.barplot(x=event_counts.index, y=event_counts.values, palette='viridis')
plt.title('Event Frequency in Event Log')
plt.xlabel('Event')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
